```
import os
import pandas as pd
```

```
data = "./tianchi_datasets"
csv_list = os.listdir(data)
csv_list = [os.path.join(i,os.listdir(os.path.join(data,i))[0]) for i in csv_list]
csv_data = []
```

```
for i in csv_list:
  print(i)
  if i == 'OCNLI/OCNLI_train1128.csv':
    print(i)
    csv_data.append(pd.read_csv(os.path.join(data,i), delimiter="\t", names=["id", '句子1', '句子2', '标签'], header=0, encoding="utf"))
  else :
    csv_data.append(pd.read_csv(os.path.join(data,i), header=0, delimiter="\t", names=['id', '句子', '标签'], encoding="utf"))
```

```
csv_list = os.listdir(data)
for i in range(len(csv_data)):
  train_data = csv_data[i].sample(frac=0.8, random_state=0, axis=0)
  dev_data = csv_data[i][~csv_data[i].index.isin(train_data.index)]
  train_data.to_csv(os.path.join(data, csv_list[i], 'train.csv'), index=None, sep='\t', encoding="utf", header=False)
  dev_data.to_csv(os.path.join(data, csv_list[i], 'dev.csv'), index=None, sep='\t', encoding="utf", header=False)
```
